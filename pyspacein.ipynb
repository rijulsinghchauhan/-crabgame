```python
import pygame
pygame.font.init()

WIDTH, HEIGHT =900, 500
WIN = pygame.display.set_mode((WIDTH,HEIGHT))
pygame.display.set_caption("HAWKEYE")
anal=(255,255,255)
BRD =(97,97,97)
BLK =(0,0,0)
YEL =(255,0,255)

BORDER = pygame.Rect(WIDTH//2-5, 0, 10, HEIGHT)

HEALTH_FONT = pygame.font.SysFont('comicsan',40)
WINNER_FONT = pygame.font.SysFont('comicsan',100)

FPS = 60
vel = 3
BULLETS_VEL = 7
MAX_BULLETS = 3
cwidth,cheight = 55,55
YELLOW_HIT = pygame.USEREVENT + 1
BLACK_HIT = pygame.USEREVENT + 2
yellow_1 = pygame.image.load('yellow.png')
yellow1 = pygame.transform.rotate(pygame.transform.scale(yellow_1,(cwidth,cheight)),270)
black_2 = pygame.image.load('black.png')
black2 = pygame.transform.rotate(pygame.transform.scale(black_2,(cwidth,cheight)),90)
SPACE = pygame.transform.scale(pygame.image.load('space.png'),(WIDTH, HEIGHT))

def draw_window(black, yellow, yellow_bullets ,black_bullets ,black_health , yellow_health):
    WIN.fill(anal)
    WIN.blit(SPACE,(0,0))
    black_health_text = HEALTH_FONT.render("Health: " +str(black_health), 1, anal)
    yellow_health_text = HEALTH_FONT.render("Health: " +str(yellow_health), 1, anal)
    WIN.blit(black_health_text,(WIDTH -black_health_text.get_width()-10, 10))
    WIN.blit(yellow_health_text,(10, 10))
    pygame.draw.rect(WIN,BRD,BORDER)
    WIN.blit(yellow1,(yellow.x,yellow.y))
    WIN.blit(black2,(black.x,black.y))
    
    for bullet in black_bullets:
        pygame.draw.rect(WIN, BLK, bullet)
    
    for bullet in yellow_bullets:
        pygame.draw.rect(WIN, YEL, bullet)    
    pygame.display.update()

def yellow_key_move(ky_prs, yellow):
        if ky_prs[pygame.K_a] and yellow.x >0: #L
            yellow.x -= vel
        if ky_prs[pygame.K_d] and yellow.x + yellow.width <BORDER.x: #R
            yellow.x += vel
        if ky_prs[pygame.K_w] and yellow.y > 0: #U
            yellow.y -= vel
        if ky_prs[pygame.K_s] and yellow.y + yellow.height<HEIGHT: #D
            yellow.y += vel
            
def black_key_move(ky_prs, black):
        if ky_prs[pygame.K_LEFT] and black.x + vel>BORDER.x+ BORDER.width : #L
            black.x -= vel
        if ky_prs[pygame.K_RIGHT] and black.x + black.width < WIDTH : #R
            black.x += vel
        if ky_prs[pygame.K_DOWN] and black.y + black.height < HEIGHT: #D
            black.y += vel
        if ky_prs[pygame.K_UP] and black.y > 0: #U
            black.y -= vel

            
            
def handle_bullets(yellow_bullets,black_bullets,yellow,black):
    for bullet in yellow_bullets:
        bullet.x += BULLETS_VEL
        if black.colliderect(bullet):
            pygame.event.post(pygame.event.Event(BLACK_HIT))
            yellow_bullets.remove(bullet)
        elif bullet.x > WIDTH:
            yellow_bullets.remove(bullet)
            
    for bullet in black_bullets:
        bullet.x -= BULLETS_VEL
        if yellow.colliderect(bullet):
            pygame.event.post(pygame.event.Event(YELLOW_HIT))
            black_bullets.remove(bullet)
        elif bullet.x < 0:
             black_bullets.remove(bullet)

                
def draw_winner(text):
    draw_text = WINNER_FONT.render(text,1,anal)
    WIN.blit(draw_text,(WIDTH/2-draw_text.get_width()/2,HEIGHT/2-draw_text.get_height()/2))
    pygame.display.update
    pygame.time.delay(5000)
            
def main():
    yellow = pygame.Rect(100,100,cwidth,cheight)
    black = pygame.Rect(700,100,cwidth,cheight)
    
    black_bullets = []
    yellow_bullets = []
    
    black_health = 10
    yellow_health = 10
    clock = pygame.time.Clock()
    run = True
    while run:
        clock.tick(FPS)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
                pygame.quit()
                
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LCTRL and len(yellow_bullets) < MAX_BULLETS:
                    bullet = pygame.Rect(
                        yellow.x + yellow.width, yellow.y + yellow.height//2 - 2, 10, 5)
                    yellow_bullets.append(bullet)
                
                if event.key == pygame.K_RCTRL and len(black_bullets) < MAX_BULLETS:
                    bullet = pygame.Rect(
                        black.x, black.y + black.height//2 - 2, 10, 5)
                    black_bullets.append(bullet)
                    
            if event.type == BLACK_HIT:
                black_health -= 1
            
            if event.type == YELLOW_HIT:
                yellow_health -= 1
                
        winner_text = ""
        if black_health <= 0:
            winner_text = "Yellow wins!"
        
        if yellow_health <= 0:
            winner_text = "Black wins!"

        if winner_text != "":
            draw_winner(winner_text)
            break
           
        
        
        ky_prs = pygame.key.get_pressed()
        yellow_key_move(ky_prs, yellow)
        black_key_move(ky_prs, black)
        handle_bullets(yellow_bullets,black_bullets,yellow,black)
        draw_window(black, yellow, yellow_bullets ,black_bullets ,black_health , yellow_health)
                
    main()
if __name__ == "__main__":
    main()
```
